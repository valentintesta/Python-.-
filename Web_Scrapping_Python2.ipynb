```python
#Importar librerias
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import requests
```

```python
#Leemos la dirección de ML y la parseamos (tener toda la página en un objeto)

url_parcial = 'https://listado.mercadolibre.com.ar/iphones#D[A:iphones]'
response= requests.get(url_parcial)
soup = BeautifulSoup(response.content,"html.parser")
```

```python
nombre = []
precio = []
condicion = []

celulares = soup.find_all("li", class_="ui-search-layout__item")

for celular in celulares:
    precio.append(celular.find(class_='price-tag-fraction'))
    nombre.append(celular.find(class_='ui-search-item__title').text)
    condicion.append(celular.find(class_='ui-search-item__group__element ui-search-item__details'))
```

```python
# Crear el DataFrame
df = pd.DataFrame()
df['Celulares'] = nombre
df['Precio'] = precio
df['Condicion'] = condicion
df['Precio'] = df['Precio'].astype(float)
```

```python
df
```

```python
#Construimos un dataframe para cada modelo para poder obtener el precio promedio de cada uno

df_iPhone_13 = df[df.Celulares.str.contains('iPhone 13')]
df_iPhone_12 = df[df.Celulares.str.contains('iPhone 12')]
df_iPhone_11 = df[df.Celulares.str.contains('iPhone 11')]
df_iPhone_X = df[df.Celulares.str.contains('iPhone X')]
```

```python
#Construimos una fila con el día actual y los precios promedio de cada modelo

precios_promedios = [pd.datetime.today().strftime('%Y-%m-%d'),df_iPhone_13['Precio'].mean(),df_iPhone_12['Precio'].mean(),df_iPhone_11['Precio'].mean(),df_iPhone_X['Precio'].mean()]
```

```python
#Construimos el dataframe final que contendrá los precios promoedio de cada día

df_final =pd.DataFrame(columns = ['Date', 'iPhone 13', 'iPhone 12', 'iPhone 11', 'iPhone X'])
```

```python
#Agregamos la fila que construimos con el día y los precios

df_final.loc[0] = precios_promedios
```
